```python
from unsloth import FastLanguageModel
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, f1_score,
                             classification_report, confusion_matrix)
from transformers import TrainingArguments
from trl import SFTTrainer
import torch
import warnings, re
warnings.filterwarnings("ignore")

print("вң… All libraries imported successfully!")
print(f"рҹ–ҘпёҸ  GPU: {torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'CPU only'}")
print(f"рҹ”ў  CUDA available: {torch.cuda.is_available()}")
```

## рҹ“Ӯ Step 1 вҖ” Load & Explore Dataset

**Dataset:** `EXt1/Thai-True-Fake-News`  
- **Unnamed: 0** вҖ” Row ID  
- **Title** вҖ” Thai news headline (input)  
- **Verification_Status** вҖ” `аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ` (real) or `аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ` (fake) (label)


```python
# а№Ӯаё«аёҘаё” Dataset аёҲаёІаёҒ Hugging Face
dataset = load_dataset("EXt1/Thai-True-Fake-News")
df = dataset["train"].to_pandas()

print(f"рҹ“Ҡ Dataset Shape: {df.shape}")
print(f"\nрҹҸ·пёҸ  Columns: {df.columns.tolist()}")
print(f"\nрҹ“Ӣ Sample rows:")
df.head(10)
```

```python
# аё•аёЈаё§аёҲаёӘаёӯаёҡ Label Distribution
print("=" * 50)
print("рҹ“Ҡ Label Distribution")
print("=" * 50)
label_counts = df["Verification_Status"].value_counts()
print(label_counts)
print(f"\nрҹ”ө аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ (Real) : {label_counts.get('аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ', 0):,}")
print(f"рҹ”ҙ аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ (Fake) : {label_counts.get('аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ', 0):,}")

# Plot distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Bar chart
axes[0].bar(label_counts.index, label_counts.values,
            color=["#F44336","#2196F3"], edgecolor="black", linewidth=0.8)
axes[0].set_title("Label Distribution (Bar Chart)", fontsize=13, fontweight="bold")
axes[0].set_xlabel("Label")
axes[0].set_ylabel("Count")
for i, v in enumerate(label_counts.values):
    axes[0].text(i, v + 10, str(v), ha="center", fontsize=11, fontweight="bold")

# Pie chart
axes[1].pie(label_counts.values, labels=label_counts.index,
            autopct="%1.1f%%", colors=[ "#F44336","#2196F3"], startangle=90)
axes[1].set_title("Label Distribution (Pie Chart)", fontsize=13, fontweight="bold")

plt.tight_layout()
plt.savefig("label_distribution.png", dpi=120, bbox_inches="tight")
plt.show()
print("\nвң… Distribution plot saved!")
```

```python
# аё•аёЈаё§аёҲаёӘаёӯаёҡ Text Length
df["title_length"] = df["Title"].apply(lambda x: len(str(x)))

print("рҹ“Ҹ Title Length Statistics:")
print(df.groupby("Verification_Status")["title_length"].describe().round(2))

plt.figure(figsize=(10, 4))
for label, color in [("аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ", "#2196F3"), ("аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ", "#F44336")]:
    subset = df[df["Verification_Status"] == label]["title_length"]
    plt.hist(subset, bins=40, alpha=0.6, label=label, color=color)
plt.xlabel("Title Length (characters)")
plt.ylabel("Count")
plt.title("Distribution of Title Length by Label")
plt.legend()
plt.tight_layout()
plt.savefig("title_length_distribution.png", dpi=120, bbox_inches="tight")
plt.show()
```

## рҹ”§ Step 2 вҖ” Pre-process & Split Data

Split ratio: **70% train / 15% validation / 15% test**  
Stratified split ensures equal label balance across all sets.


```python
# аё—аёіаё„аё§аёІаёЎаёӘаё°аёӯаёІаё”аёӮа№үаёӯаёЎаё№аёҘ
df = df.dropna(subset=["Title", "Verification_Status"]).reset_index(drop=True)
df["Title"] = df["Title"].astype(str).str.strip()

# аёҒаёЈаёӯаёҮа№ҖаёүаёһаёІаё° label аё—аёөа№Ҳаё–аё№аёҒаё•а№үаёӯаёҮ
valid_labels = ["аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ", "аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ"]
df = df[df["Verification_Status"].isin(valid_labels)].reset_index(drop=True)

print(f"вң… Clean dataset: {len(df):,} rows")

# а№Ғаёҡа№ҲаёҮ train / val / test  (70 / 15 / 15)
train_df, temp_df = train_test_split(
    df, test_size=0.30, random_state=42, stratify=df["Verification_Status"])
val_df, test_df = train_test_split(
    temp_df, test_size=0.50, random_state=42, stratify=temp_df["Verification_Status"])

print(f"\nрҹ“Ұ Split sizes:")
print(f"  рҹ”№ Train : {len(train_df):,} rows")
print(f"  рҹ”№ Val   : {len(val_df):,} rows")
print(f"  рҹ”№ Test  : {len(test_df):,} rows")

# Verify label balance
for name, d in [("Train", train_df), ("Val", val_df), ("Test", test_df)]:
    counts = d["Verification_Status"].value_counts()
    print(f"\n  {name}: аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ={counts.get('аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ',0)}, аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ={counts.get('аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ',0)}")
```

## рҹӨ– Step 3 вҖ” Load Qwen2.5-7B with Unsloth

- **Model:** `unsloth/Qwen2.5-7B-Instruct-bnb-4bit`  
- **4-bit Quantization (QLoRA):** аёҘаё”аёӮаёҷаёІаё” VRAM аёҲаёІаёҒ ~28GB вҶ’ ~6GB  
- **LoRA rank = 16:** Fine-tune а№ҖаёүаёһаёІаё° adapter layers а№Ғаё—аёҷаё—аёұа№үаёҮ model  


```python
# аёҒаёіаё«аёҷаё” config
max_seq_length = 512   # аё„аё§аёІаёЎаёўаёІаё§ token аёӘаё№аёҮаёӘаёёаё” (аёӮа№ҲаёІаё§аёӘаёұа№үаёҷ вҶ’ 512 аёһаёӯ)
dtype = None           # Auto-detect
load_in_4bit = True    # QLoRA - аёӣаёЈаё°аё«аёўаёұаё” VRAM

# а№Ӯаё«аёҘаё” Model + Tokenizer аё”а№үаё§аёў Unsloth
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name    = "unsloth/Qwen2.5-7B-Instruct-bnb-4bit",
    max_seq_length = max_seq_length,
    dtype          = dtype,
    load_in_4bit   = load_in_4bit,
)

print("вң… Model loaded successfully!")
print(f"рҹ“җ Model type: {type(model).__name__}")
```

```python
# а№Җаёһаёҙа№ҲаёЎ LoRA Adapters
model = FastLanguageModel.get_peft_model(
    model,
    r              = 16,      # LoRA rank (аёўаёҙа№ҲаёҮаёӘаё№аёҮ = а№ҖаёЈаёөаёўаёҷаёЈаё№а№үаёЎаёІаёҒаёӮаё¶а№үаёҷ а№Ғаё•а№Ҳа№ғаёҠа№ү VRAM аёЎаёІаёҒаёӮаё¶а№үаёҷ)
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj"],
    lora_alpha     = 16,
    lora_dropout   = 0.05,
    bias           = "none",
    use_gradient_checkpointing = "unsloth",  # аёӣаёЈаё°аё«аёўаёұаё” VRAM а№Җаёһаёҙа№ҲаёЎа№Җаё•аёҙаёЎ
    random_state   = 42,
)

# аёҷаёұаёҡаёҲаёіаёҷаё§аёҷ parameter
total_params     = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\nрҹ“Ҡ Model Parameters:")
print(f"  Total     : {total_params:,}")
print(f"  Trainable : {trainable_params:,} ({100*trainable_params/total_params:.2f}%)")
print(f"\nвң… LoRA adapters added!")
```

## рҹ“қ Step 4 вҖ” Prompt Template (Alpaca Format)

а№ғаёҠа№ү **Alpaca-style prompt** а№Җаёһаё·а№ҲаёӯаёҡаёӯаёҒа№ғаё«а№ү model аёЈаё№а№үаё§а№ҲаёІ:
- Task аё„аё·аёӯаёӯаё°а№„аёЈ (instruction)
- Input аё„аё·аёӯаёӯаё°а№„аёЈ (title аёӮа№ҲаёІаё§)
- Output аё—аёөа№Ҳаё•а№үаёӯаёҮаёҒаёІаёЈ (аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ / аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ)


```python
# Alpaca-style prompt template аёӘаёіаё«аёЈаёұаёҡ Fake News Detection
alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
аё„аёёаё“а№Җаёӣа№Үаёҷ AI аёңаё№а№үа№ҖаёҠаёөа№Ҳаёўаё§аёҠаёІаёҚаё”а№үаёІаёҷаёҒаёІаёЈаё•аёЈаё§аёҲаёӘаёӯаёҡаёӮа№ҲаёІаё§аё аёІаё©аёІа№„аё—аёў аёҒаёЈаёёаё“аёІаё§аёҙа№Җаё„аёЈаёІаё°аё«а№Ңаё«аёұаё§аёӮа№үаёӯаёӮа№ҲаёІаё§аё•а№Ҳаёӯа№„аёӣаёҷаёөа№үа№ҒаёҘаё°аё•аёӯаёҡаё§а№ҲаёІа№Җаёӣа№Үаёҷ "аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ" аё«аёЈаё·аёӯ "аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ" а№Җаё—а№ҲаёІаёҷаёұа№үаёҷ

### Input:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token  # аё•а№үаёӯаёҮа№ғаёӘа№Ҳ EOS аё—а№үаёІаёў

def format_prompt(title, label=""):
    """а№ҒаёӣаёҘаёҮ row а№Җаёӣа№Үаёҷ prompt string"""
    return alpaca_prompt.format(title, label) + (EOS_TOKEN if label else "")

# а№ҒаёӘаё”аёҮаё•аёұаё§аёӯаёўа№ҲаёІаёҮ prompt
example_row = train_df.iloc[0]
print("рҹ“„ Example Prompt:\n")
print(format_prompt(example_row["Title"], example_row["Verification_Status"]))
```

```python
# а№ҒаёӣаёҘаёҮ DataFrame а№Җаёӣа№Үаёҷ Hugging Face Dataset format
from datasets import Dataset

def prepare_dataset(dataframe):
    """а№ҒаёӣаёҘаёҮ DataFrame вҶ’ HF Dataset аё—аёөа№ҲаёһаёЈа№үаёӯаёЎ train"""
    texts = []
    for _, row in dataframe.iterrows():
        text = format_prompt(row["Title"], row["Verification_Status"])
        texts.append({"text": text})
    return Dataset.from_list(texts)

train_dataset = prepare_dataset(train_df)
val_dataset   = prepare_dataset(val_df)

print(f"вң… Train dataset: {len(train_dataset):,} samples")
print(f"вң… Val   dataset: {len(val_dataset):,} samples")
print(f"\nрҹ”Қ Sample entry:")
print(train_dataset[0]["text"][:300])
```

## рҹҸӢпёҸ Step 5 вҖ” Fine-tune with SFTTrainer

**Hyperparameters:**
| Parameter | Value |
|---|---|
| Epochs | 2 |
| Batch size | 2 |
| Gradient accumulation | 4 (effective batch = 8) |
| Learning rate | 2e-4 |
| LR Scheduler | cosine |
| Optimizer | adamw_8bit |


```python
# аёҒаёіаё«аёҷаё” Training Arguments
training_args = TrainingArguments(
    output_dir              = "./qwen_fakenews_output",
    num_train_epochs        = 3,
    per_device_train_batch_size  = 2,
    per_device_eval_batch_size   = 2,
    gradient_accumulation_steps  = 4,
    warmup_ratio            = 0.05,
    learning_rate           = 2e-4,
    lr_scheduler_type       = "cosine",
    optim                   = "adamw_8bit",
    fp16                    = not torch.cuda.is_bf16_supported(),
    bf16                    = torch.cuda.is_bf16_supported(),
    logging_steps           = 10,
    eval_strategy           = "epoch",   # вҶҗ changed from evaluation_strategy
    save_strategy           = "epoch",
    load_best_model_at_end  = True,
    metric_for_best_model   = "eval_loss",
    seed                    = 42,
    report_to               = "none",
)

# аёӘаёЈа№үаёІаёҮ SFTTrainer
trainer = SFTTrainer(
    model              = model,
    tokenizer          = tokenizer,
    train_dataset      = train_dataset,
    eval_dataset       = val_dataset,
    dataset_text_field = "text",
    max_seq_length     = max_seq_length,
    dataset_num_proc   = 2,
    args               = training_args,
)

print("вң… Trainer created!")
print(f"рҹ“Ң Training on {len(train_dataset):,} samples")
print(f"рҹ“Ң Validating on {len(val_dataset):,} samples")
```

```python
# рҹҡҖ а№ҖаёЈаёҙа№ҲаёЎ Training!
print("рҹҡҖ Starting Fine-tuning...\n")
trainer_stats = trainer.train()

print("\nвң… Training Complete!")
print(f"вҸұпёҸ  Total training time: {trainer_stats.metrics['train_runtime']:.1f} seconds")
print(f"рҹ“ү Final train loss   : {trainer_stats.metrics['train_loss']:.4f}")
```

## рҹ“Ҳ Step 6 вҖ” Plot Training & Validation Loss

```python
# аё”аё¶аёҮ training log history
log_history = trainer.state.log_history

# а№ҒаёўаёҒ train loss а№ҒаёҘаё° eval loss
train_logs = [x for x in log_history if "loss" in x and "eval_loss" not in x]
eval_logs  = [x for x in log_history if "eval_loss" in x]

train_steps  = [x["step"] for x in train_logs]
train_losses = [x["loss"] for x in train_logs]
eval_steps   = [x["step"] for x in eval_logs]
eval_losses  = [x["eval_loss"] for x in eval_logs]

# Plot
plt.figure(figsize=(10, 5))
plt.plot(train_steps, train_losses, label="Train Loss", color="#2196F3", linewidth=2)
if eval_losses:
    plt.plot(eval_steps, eval_losses, label="Val Loss", color="#F44336",
             linewidth=2, linestyle="--", marker="o", markersize=7)
plt.xlabel("Training Steps")
plt.ylabel("Loss")
plt.title("Training & Validation Loss Curve", fontsize=14, fontweight="bold")
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("training_loss.png", dpi=120, bbox_inches="tight")
plt.show()
print("вң… Loss plot saved!")
```

## рҹ“Ҡ Step 7 вҖ” Evaluate on Test Set

Metrics: **Accuracy**, **F1 Score** (weighted), **Classification Report**, **Confusion Matrix**


```python
# аёҹаёұаёҮаёҒа№ҢаёҠаёұаёҷ inference
FastLanguageModel.for_inference(model)  # а№Җаёӣаёҙаё” optimized inference mode

def predict_label(title: str) -> str:
    """аёЈаёұаёҡаё«аёұаё§аёӮа№үаёӯаёӮа№ҲаёІаё§ вҶ’ аё„аё·аёҷаё„а№ҲаёІ 'аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ' аё«аёЈаё·аёӯ 'аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ'"""
    prompt = alpaca_prompt.format(title, "")   # а№„аёЎа№Ҳа№ғаёӘа№Ҳ label (а№ғаё«а№ү model аё—аёІаёў)
    inputs = tokenizer([prompt], return_tensors="pt").to("cuda")

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens = 10,
            temperature    = 0.1,
            do_sample      = True,
            pad_token_id   = tokenizer.eos_token_id,
        )

    # Decode а№ҖаёүаёһаёІаё° token аё—аёөа№Ҳ generate а№ғаё«аёЎа№Ҳ
    generated = tokenizer.decode(
        outputs[0][inputs["input_ids"].shape[1]:], skip_special_tokens=True
    ).strip()

    # Map output вҶ’ clean label
    if "аёҲаёЈаёҙаёҮ" in generated:
        return "аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ"
    elif "аёӣаёҘаёӯаёЎ" in generated:
        return "аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ"
    else:
        return generated   # return raw аё–а№үаёІ model аё•аёӯаёҡаёңаёҙаё”аёЈаё№аёӣа№Ғаёҡаёҡ

# аё—аё”аёӘаёӯаёҡаёҒаёұаёҡаё•аёұаё§аёӯаёўа№ҲаёІаёҮ
sample_title = test_df.iloc[0]["Title"]
sample_label = test_df.iloc[0]["Verification_Status"]
pred = predict_label(sample_title)
print(f"рҹ“° Title  : {sample_title}")
print(f"рҹҸ·пёҸ  True   : {sample_label}")
print(f"рҹӨ– Predict: {pred}")
```

```python
# аёӣаёЈаё°а№ҖаёЎаёҙаёҷаёңаёҘаёҡаёҷ Test Set аё—аёұа№үаёҮаё«аёЎаё”
from tqdm.auto import tqdm

print("рҹ”„ Running inference on test set...\n")
y_true = test_df["Verification_Status"].tolist()
y_pred = []

for title in tqdm(test_df["Title"].tolist(), desc="Predicting"):
    y_pred.append(predict_label(title))

# аё„аёіаёҷаё§аё“ metrics
accuracy = accuracy_score(y_true, y_pred)
f1       = f1_score(y_true, y_pred, average="weighted",
                    labels=["аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ", "аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ"])

print(f"\n{'='*55}")
print("рҹ“Ҡ EVALUATION RESULTS")
print(f"{'='*55}")
print(f"вң… Accuracy  : {accuracy*100:.2f}%")
print(f"вң… F1 Score  : {f1*100:.2f}%  (weighted)")
print(f"\nрҹ“Ӣ Classification Report:")
print(classification_report(y_true, y_pred,
                             target_names=["аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ", "аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ"],
                             zero_division=0))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_true, y_pred, labels=["аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ", "аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ"])

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ", "аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ"],
            yticklabels=["аёӮа№ҲаёІаё§аёҲаёЈаёҙаёҮ", "аёӮа№ҲаёІаё§аёӣаёҘаёӯаёЎ"],
            linewidths=0.5, annot_kws={"size": 14})
plt.xlabel("Predicted Label", fontsize=12)
plt.ylabel("True Label", fontsize=12)
plt.title("Confusion Matrix вҖ” Thai Fake News Detector", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("confusion_matrix.png", dpi=120, bbox_inches="tight")
plt.show()
print("вң… Confusion matrix saved!")
```
